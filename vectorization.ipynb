Then, **identify the two reviews that are the most "similar" based upon cosine similarity**. (7 pts)

```python
import pandas as pd
with open("poor_amazon_toy_reviews.txt","r") as poorfile:
    poorlines=poorfile.readlines()
with open("good_amazon_toy_reviews.txt","r") as goodfile:
    goodlines=goodfile.readlines()
poorwords=open("poor_amazon_toy_reviews.txt","r").read()
goodwords=open("good_amazon_toy_reviews.txt","r").read()
```

```python
totalines=poorlines+goodlines
```

### Tokenize all words in comments

```python
from nltk.tokenize import word_tokenize
poorword = word_tokenize(poorwords)
goodword = word_tokenize(goodwords)
totalword=poorword+goodword
```

### Lemmatization

```python
from nltk.stem import WordNetLemmatizer
lemmatizer = WordNetLemmatizer()
words=[]
for word in totalword:
    words.append(lemmatizer.lemmatize(word))
```

### Create word count dataframe and find top 500 words with the highest frequency

```python
from sklearn.feature_extraction.text import CountVectorizer
vectorizer=CountVectorizer(stop_words="english")
vectorizer.fit(words)
vector=vectorizer.transform(words)
sum=vector.sum(axis=0)
freq=[(word,sum[0,idx]) for word, idx in vectorizer.vocabulary_.items()]
freq=sorted(freq, key=lambda x:x[1],reverse=True)
df=freq[:500] 
```

```python
highword=[]
for word in df:
    highword.append(word[0])
```

### Create correlation matrix for top 500 words

```python
vectorizer=CountVectorizer()
result=vectorizer.fit_transform(highword)
count=pd.DataFrame(result.toarray(),columns=vectorizer.get_feature_names())
count.corr().to_csv("hwcorr_matrix.csv")
count.corr()
```

### Random select 200 comments from the total comments

```python
import random
random.seed(86)
```

```python
X = vectorizer.fit_transform(totalines)
df=pd.DataFrame(X.toarray(),columns=vectorizer.get_feature_names())
df = df.sample(200)
```

```python
df
```

### Find 2 comments with the highest cosine similarity from above 200 comments

```python
from numpy import dot
from numpy.linalg import norm
result=[]    
for i in range(1,df.shape[0]):
    for k in range(1,df.shape[0]):
        if i!=k:
            x=df.iloc[i]
            y=df.iloc[k]
            dis=dot(x, y)/(norm(x)*norm(y))
            row=[i,k,dis]
            result.append(row)
        
similarity = pd.DataFrame(result,columns=['sen1','sen2','dist'])       
```

```python
similarity.loc[similarity['dist'].idxmax()]
```
