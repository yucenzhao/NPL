```python
import pandas as pd
poorfile=open("poor_amazon_toy_reviews.txt","r")
goodfile=open("good_amazon_toy_reviews.txt","r")
poorfile.seek(0)
poorlines=poorfile.readlines()
print(len(poorlines))
```

```python
goodfile.seek(0)
goodlines=goodfile.readlines()
print(len(goodlines))
```

```python
from collections import Counter
def count_words(lines,delimiter=" "):
    words=Counter()
    
    for line in lines:
        for word in line.split(delimiter):
            words[word]+=1  

    return(words) 
poor_count=count_words(poorlines)
```

```python
poor_count
```

```python
good_count=count_words(goodlines)
```
